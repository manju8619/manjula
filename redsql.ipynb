```python
import pandas as pd
df = pd.read_csv('concatenated_file.csv').set_index('Unnamed: 0')
```

```python
import numpy as np

# Define the positions (ratings) and their probabilities
positions = [1, 2, 3, 4, 5]
probabilities = [0.10, 0.20, 0.50, 0.10, 0.10]

# Generate random ratings for all rows in the DataFrame
ratings = np.random.choice(positions, size=df.shape[0], p=probabilities)

# Assign the generated ratings to the 'star_rating' column
df["star_rating"] = ratings
```

```python
mycursor.execute("create table redbus.databases (bus_route_name VARCHAR(25),bus_type VARCHAR(25),departing_time INT(10),duration INT(25),reaching_time INT(25),star_rating INT(10),price VARCHAR(25),seat_availability VARCHAR(25))")
```
